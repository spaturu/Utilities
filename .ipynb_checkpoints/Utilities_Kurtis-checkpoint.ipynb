## Visualize Example Time Series for Residential Lines

```python
df_res = data[data['AccountType'] == 'RESIDENTIAL'].copy()
df_res = df_res[['Date','TimeStamp','Value1']]
```

```python
df_res.sort_values(['Date','TimeStamp'])
```

```python
df_res['DateTime'] = df_res.apply(lambda r: r['Date'] + ' ' + r['TimeStamp'], axis = 1)
df_res['DateTime'] = pd.to_datetime(df_res['DateTime'])
```

```python
df_res
```

```python
gb_res = df_res.groupby([df_res['DateTime'].dt.year, df_res['DateTime'].dt.month, df_res['DateTime'].dt.day, df_res['DateTime'].dt.hour, df_res['DateTime'].dt.minute]) 
```

```python
groups = dict(list(gb_res))
print(len(groups))
```

```python
dates = list(groups.keys())
```

```python
X_datetime = np.asarray([datetime(*dates[i]) for i in range(len(dates))])
```

```python
X_datetime.shape
```

```python
X_values = np.asarray([gb_res.get_group(dates[i]).mean()[0] for i in range(len(dates))])
X_values.shape
```

```python
# sort by datetime
p = X_datetime.argsort()
X_datetime = X_datetime[p]
X_values = X_values[p]
```

## Create Training/Testing Set for Time Series Regression

```python
n = X_values.shape[0]

# 80-20 split
X_train = X_values[:int(n*.8)] 
X_test = X_values[int(n*.8):] 
```

```python
mu = np.mean(X_train)
sigma = np.std(X_train)
```

```python
X_train = (X_train-mu)/sigma
X_test = (X_test-mu)/sigma
```

```python
def transform_inputs(X, time_window):
    
    X_inputs = []
    y = []
    
    for i in range(len(X)-time_window-1):
        
        X_inputs.append(X[i:i+time_window])
        y.append(X[i+time_window])
    
    return np.asarray(X_inputs),np.asarray(y)
```

```python
time_window = 100
X_train_inputs,y_train = transform_inputs(X_train[:int(X_train.shape[0]*.8)],time_window)
X_val_inputs,y_val = transform_inputs(X_train[int(X_train.shape[0]*.8):],time_window)
X_test_inputs,y_test = transform_inputs(X_test,time_window)
```

```python
print(X_train_inputs.shape, y_train.shape)
print(X_val_inputs.shape, y_val.shape)
print(X_test_inputs.shape, y_test.shape)
```

## Test using Multilayer Perceptron ('Normal' Feedforward ANN)

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.optimizers import RMSprop
```

```python
model = Sequential()
model.add(Dense(64, activation = 'relu', input_dim=time_window))
model.add(Dropout(0.2))
model.add(Dense(1, activation = 'linear'))
model.compile(loss='mean_squared_error',
              optimizer='sgd')
```

```python
batch_size = 100
epochs = 100
history = model.fit(X_train_inputs, y_train,
                    batch_size = batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_val_inputs, y_val))
```

```python
score = model.evaluate(X_test_inputs, y_test, verbose=0)
```

## Create LSTM

```python
trainX = np.reshape(X_train_inputs, (X_train_inputs.shape[0], 1, X_train_inputs.shape[1]))
valX = np.reshape(X_val_inputs, (X_val_inputs.shape[0], 1, X_val_inputs.shape[1]))
```

```python
trainX
```

```python
from keras.layers import LSTM
rnn_model=None
rnn_model = Sequential()
rnn_model.add(LSTM(128, input_shape = (1,time_window), return_sequences = False))
rnn_model.add(Dropout(0.5))
rnn_model.add(Dense(1, activation = 'linear'))
rnn_model.compile(loss="mean_squared_error", optimizer="rmsprop")
```

```python
rnn_model.fit(trainX, y_train, epochs = 100, batch_size = 100, verbose = 1)
```

```python
y_pred_train = rnn_model.predict(trainX)
```

```python
y_val_train = rnn_model.predict(valX)
```
